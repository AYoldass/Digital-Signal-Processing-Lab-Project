Now let's change the sampling frequency to $f_s=20 $ Hz. Repeat the previous parts (with sampling and reconstruction). Does it resemble the original signal? Is this sampling frequency satisfies the Nyquist rate? 

```python
# Write your code here
f=10
fs=20
t=np.arange(0,0.5+1/fs,1/fs)
x=np.cos(2*np.pi*f*t)
plt.figure()
plt.title("Recon")
plt.ylabel("Amp")
plt.xlabel("t")
plt.plot(t,x)
plt.figure()
plt.title("Samp")
plt.ylabel("Amp")
plt.xlabel("t")
plt.stem(t,x)
plt.show()
```

In this question, we did not observe aliasing since our sampling frequency was equal to 2f1, so we could not obtain a smooth sinusodial signal, but since 2*f1=fs, the original signal was saved.

```python
# Write your code here
samp = 100;   # sampling rate in Hz
t3 = np.arange(0,0.583,1/samp)  # time (1s of data)
N = len(t3)                       # store the number of time points
x5 = np.cos(20*np.pi*t3)   # the signal equation
plt.plot(t3,x5,'.-')
plt.show()
plt.title("Sampled")
plt.xlabel("t")
plt.ylabel("Signal")
plt.stem(t3,x5)
plt.show()
```

In this part, we will see the effects of aliasing and undersampling.

Consider the two sinusoidal signals:
$ x_1(t)=cos(2 \pi f_1 t)$ and $ x_2(t)=cos(2 \pi (f_1+f_2) t)$ where $f_2>2 f_1$. We will show that we cannot differentiate these signals if we sample them using the sampling frequency $f_s=f_2$. 

First plot the two signals when $f_1=10$ Hz and $f_2=100$ Hz in the same time interval (e.g between 0 and 0.2) on the same figure. You should plot signals in different colors and add legend to the figure. 

```python
# Write your code here
t2=np.arange(0,0.2,0.001)
x2=np.cos(20*np.pi*t2)
x3=np.cos(220*np.pi*t2)
plt.plot(t2, x2, 'b-', label="f1=10hz")
plt.plot(t2, x3, 'k-',label="f2=110hz") 
plt.legend(loc="lower right")
plt.ylabel("signals")
plt.xlabel("t")
plt.show()
```

Since we get the sampling frequency high, the signals can be distinguished more clearly. I suppressed the two signals on the same graph with the plt.legend function.

As you have done in the first part, take samples from two signals, visualize these discrete-time signals $x_1(nT_s)$ and $x_2(nT_s)$ with plt.stem() function and try to reconstruct the original signal so plot interpolated signals with plt.plot().

```python
# Write your code here
f1=10
f2=100
t1=np.arange(0,0.2+1/100,1/100)
x1=np.cos(2*np.pi*f1*t1)
x2=np.cos(2*np.pi*(f1+f2)*t1)

plt.figure()
plt.title("Reconstracted")
plt.xlabel("t")
plt.ylabel("Signal")
plt.plot(t1,x1,'k')

plt.figure()
plt.title("Sampled")
plt.xlabel("t")
plt.ylabel("Signal")
plt.stem(t1,x1)
plt.show()

plt.title("Reconstracted2")
plt.xlabel("t")
plt.ylabel("Signal")
plt.plot(t1,x2,'k')

plt.figure()
plt.title("Sampled2")
plt.xlabel("t")
plt.ylabel("Signal")
plt.stem(t1,x2)
plt.show()
```

First of all, I determined the frequency of the second signal to be more than twice the frequency of the first signal (f2>2f1).I specified the sampling frequency and value range in the np block.Aliasing occurs because the sum of our two frequencies is less than our sampling frequency. Since our sampling frequency is equal to the frequency we determined in the first graph, alesing event occurs and therefore the graphs overlap. Since the signals are already aliasing, we examine the same signal when we reconstruct it.
